```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv('adult.data.csv')
df.head()
```

```python
race_count = df['race'].value_counts()
race_count
```

```python
average_age_men = df[df['sex'] == "Male"]['age'].mean()
average_age_men
```

```python
bachelors = df[df['education'] == 'Bachelors'].shape[0]/df.shape[0] * 100
bachelors
```

```python
higher_education = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
percentage_higher_education = higher_education.shape[0]/df.shape[0] * 100
percentage_higher_education
```

```python
lower_education = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
percentage_lower_education = lower_education.shape[0]/df.shape[0] * 100
percentage_lower_education
lower_education
```

```python
high_salary = df[df['salary'] == '>50K']
high_salary
```

```python
higher_education_rich = pd.merge(high_salary, higher_education, how='inner')
higher_education_rich
```

```python
lower_education_rich = pd.merge(high_salary, lower_education, how='inner')
lower_education_rich
```

```python
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

```python
min_workers = df[df['hours-per-week'] == 1]
num_min_workers = pd.merge(min_workers, high_salary, how = 'inner').shape[0]
num_min_workers
```

```python
countries = high_salary['native-country'].value_counts()
highest_earning_country = countries.index[0]
highest_earning_country_percentage = countries.values[0]/high_salary.shape[0] * 100
```

```python
high_in_IN = high_salary[high_salary['native-country'] == 'India']
top_IN_occupation = high_in_IN['occupation'].value_counts().index[0]
top_IN_occupation
```
